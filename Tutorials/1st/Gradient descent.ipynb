```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

### 1Â° example:

#### Function to find minimum
$ f(x) = x^4 - 3 x^3 $

#### First order derivative
$ f'(x) = 4x^3 - 9 x^2 $

```python
# Function
f = lambda x: x**4 - 3 * x**3
df = lambda x: 4 * x**3 - 9 * x**2
```

```python
x = np.linspace(-1,3,1000) # 100 linearly spaced numbers
y = np.array([f(i) for i in x])

# compose plot
plt.plot(x,y) # sin(x)/x
plt.show() # show the plot
```

```python
# Starting point (can be randomly chosen)
x = 4

# Learning rate
LR = 0.01

# Iterate
for i in range(5000):
    # Gradient descent
    x = x - LR * df(x)

print(x)
```

### 2Â° example:

#### Function to find minimum
$ f(x) = 3x - log(x) $

#### First order derivative
$ f'(x) = 3 - (1 / x) $

```python
f = lambda x: 3*x - math.log(x)
df = lambda x: 3 - (1./x)
```

```python
x = np.linspace(0.01,2,1000) # 100 linearly spaced numbers
y = np.array([f(i) for i in x])

# compose plot
plt.plot(x,y) # sin(x)/x
plt.show() # show the plot
```

```python
# Starting point (can be randomly chosen)
x = 4

# Learning rate
LR = 0.01

for i in range(1000):
    x = x - LR * df(x)

print(x)
```
