#### Operations with matrices

```python
A = np.array([[1., 3., 4.], [2., 5., 7.]])
print(A)
print("\nShape of matrix A", A.shape, "\n\n")

B = np.ones_like(A)
print(B)
print("\nShape of matrix B", B.shape)

print("\nAddition")
print(A+B)

print("\nSubtraction")
print(A-B)

print("\nElement-wise multiplication")
print(A*np.zeros_like(A))

print("\nElement-wise division")
print(A/B)
```

#### Dot product 

```python
A = np.array([[1., 3., 4.], [2., 5., 7.]])
print(A)
print("\nShape of matrix A", A.shape, "\n\n")

B = np.array([[1., -3.], [2., -4.], [-1., 2.]])
print(B)
print("\nShape of matrix B", B.shape)

print("\nDot product")
print(np.dot(A, B))
```
